```python

import pandas as pd
df = pd.read_excel('Diabetes.xlsx')
df.shape
```

```python
X = df.iloc[:,0:8]
Y = df.iloc[:,8]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.transform(X)
```

<B>Keras</B> is the front-end used for designing the neural network.<br/>
<B>TensorFlow</B> is the backend automatically chooses the best way to represent the network for training and making predictions to run on your hardware, such as CPU or GPU or even distributed.

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential() #Models in Keras are defined as a sequence of layers
#Fully connected layers are defined using the Dense class
model.add(Dense(12, input_dim=8, activation='relu')) #define 1) an input layer with 8 nodes 2) a hidden layer with 12 nodes
# using rectifier activation function relu: f(x)= x+ = max(0,x)
model.add(Dense(8, activation='relu')) #second hidden layer with 8 nodes
model.add(Dense(1, activation='sigmoid')) #i.e., logistic function used on output to ensure output between 0 and 1
```

We must specify the loss function to use to evaluate a set of weights, the optimizer used to search through different weights for the network and any optional metrics we would like to collect and report during training.<br>
We will use the efficient gradient descent algorithm “adam” for no other reason that it is an efficient default. The optimizer decide the learning rate decay.<br>
In this case, we will use logarithmic loss, which for a binary classification problem is defined in Keras as “binary_crossentropy“. binary_crossentropy is the loss function used to measure the prediction error at each iteration:
<p>$$-\frac{1}{N}\sum_{i=1}^N [\color{blue}{y_i} \log(\hat{y}_i)+(1-y_i) \log(1-\hat{y}_i)]$$</p>

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

The training process will run for a fixed number of iterations through the dataset called epochs, that we must specify using the nepochs argument. We can also set the number of instances that are evaluated before a weight update in the network is performed, called the batch size and set using the batch_size argument.

```python
# Fit the model
#I was lazy and did not split into train and test sets
model.fit(X, Y, epochs=150, batch_size=10)
```

```python
# evaluate the model performance
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
